```python
import ROOT
from ROOT import TFile, TTree, TH1F, TCanvas, TAxis, TLegend, TTreeReader, TTreeReaderValue
ROOT.gStyle.SetOptStat(0)
```

```python
file = TFile.Open("/home/xyan13/Trijet/TrijetAna/TrijetAna/outputs/Res1ToRes2GluTo3Glu_M1-3000_R-0p5_ML_study.root")
tree = file.Get("Events")
```

```python
variable2use = [i.GetName() for i in tree.GetListOfBranches()]
display(variable2use)
```

```python
# variable2use = [i.GetName() for i in tree.GetListOfBranches()]
variable2use = {}
variable2use["m_jj"] = {"title": "m_{jj} [GeV]", "binning": 100, "lb": 0, "hb": 3000}
variable2use["dijet_eta"] = {"title": "#eta_{jj}", "binning": 100, "lb": 0, "hb": 2.5}
variable2use["dR_jj"] = {"title": "#delta R_{jj}", "binning": 100, "lb": 0, "hb": 10}
# variable2use["dEta_jj_pt"] = {"title": "#delta #eta_{jj}", "binning": 100, "lb": 0, "hb": 2.5}
# variable2use["dPhi_jj"] = {"title": "#delta #varphi_{jj}", "binning": 100, "lb": 0, "hb": 5}
# variable2use["jet_eta_0"] = {"title": "#eta_{j0}", "binning": 100, "lb": 0, "hb": 2.5}
# variable2use["jet_eta_1"] = {"title": "#eta_{j1}", "binning": 100, "lb": 0, "hb": 2.5}
# variable2use["jet_eta_2"] = {"title": "#eta_{j2}", "binning": 100, "lb": 0, "hb": 2.5}
# variable2use["jet_ptoverm_0"] = {"title": "p^{j0}_{T}/m_{jj}", "binning": 100, "lb": 0, "hb": 2.5}
# variable2use["jet_ptoverm_1"] = {"title": "p^{j1}_{T}/m_{jj}", "binning": 100, "lb": 0, "hb": 2.5}
# variable2use["dR_jj_j"] = {"title": "#delta R_{jj-j}", "binning": 100, "lb": 0, "hb": 2.5}
# variable2use["dEta_jj_j"] = {"title": "#delta #eta_{jj-j}", "binning": 100, "lb": 0, "hb": 2.5}
# variable2use["dPhi_jj_j"] = {"title": "#delta #varphi_{jj-j}", "binning": 100, "lb": 0, "hb": 5}
# variable2use["jet_ptoverM_0"] = {"title": "p^{j0}_{T}/M_{jjj}", "binning": 100, "lb": 0, "hb": 2.5}
# variable2use["jet_ptoverM_0"] = {"title": "p^{j1}_{T}/M_{jjj}", "binning": 100, "lb": 0, "hb": 2.5}
# variable2use["jet_ptoverM_0"] = {"title": "p^{j2}_{T}/M_{jjj}", "binning": 100, "lb": 0, "hb": 2.5}
# variable2use["dijet_ptoverM"] = {"title": "p^{jj}_{T}/M_{jjj}", "binning": 100, "lb": 0, "hb": 2.5}

histo_dic = {}
for variable, param_dict in variable2use.items():
    histo_dic[variable] = {}
    for selection in ["all","match","unmatch"]:
        print(selection, param_dict["binning"], param_dict["lb"], param_dict["hb"])
        histo_dic[variable][selection] = TH1F(f"{variable}_{selection}",f"{variable}_{selection}",param_dict["binning"], param_dict["lb"], param_dict["hb"])
display(histo_dic)
```

```python
for event in tree:
    for variable, hist_sel_dic in histo_dic.items():
        hist_sel_dic["all"].Fill(getattr(event, variable))
        if(event.gen_dijet_matched == 1):
            hist_sel_dic["match"].Fill(getattr(event, variable))
        else:
            hist_sel_dic["unmatch"].Fill(getattr(event, variable))
```

```python
for variable, hist_sel_dic in histo_dic.items():
    print(hist_sel_dic["all"].GetXaxis().GetXmin(), hist_sel_dic["match"].GetXaxis().GetXmin())
    print(hist_sel_dic["all"].GetXaxis().GetXmax(), hist_sel_dic["match"].GetXaxis().GetXmax())
#     hist_sel_dic["unmatch"] = hist_sel_dic["all"] - hist_sel_dic["match"]
```

```python
def plot(h_plot, canvas, leg, lcolor, lstyle, xtitle, dopt, hlabel):
    h_plot.SetLineWidth(2)
    h_plot.SetLineColor(lcolor)
    h_plot.SetLineStyle(lstyle)
    h_plot.Draw(dopt)
    h_plot.GetXaxis().SetTitle(xtitle)
    h_plot.GetXaxis().SetTitleOffset(1.2)
    h_plot.GetYaxis().SetTitle("Events")
    leg.AddEntry(h_plot,hlabel)
```

```python
import math
ROOT.disableJSVis()
nvar = len(histo_dic.keys())
nrow = math.ceil(nvar/3)
c = TCanvas("","",3600,900*nrow)
c.Divide(3, nrow)
ivar = 1
for variable, hist_sel_dic in histo_dic.items():
    print(hist_sel_dic["all"].GetEntries(), hist_sel_dic["match"].GetEntries(), hist_sel_dic["unmatch"].GetEntries())
    c.cd(ivar)
    leg = TLegend(0.6,0.7,0.85,0.85)
    title = variable2use[variable]["title"]
    plot(h_plot=hist_sel_dic["all"], canvas=c, leg=leg, lcolor=4, lstyle=1, xtitle=title, dopt="HIST", hlabel="all pairs")
    plot(h_plot=hist_sel_dic["match"], canvas=c, leg=leg, lcolor=2, lstyle=1, xtitle=title, dopt="HIST SAME", hlabel="matched pairs")
    plot(h_plot=hist_sel_dic["unmatch"], canvas=c, leg=leg, lcolor=ROOT.kGreen-2, lstyle=1, xtitle=title, dopt="HIST SAME", hlabel="aunmatched pairs")
    leg.Draw()
    ivar+=1
c.Draw()
```
