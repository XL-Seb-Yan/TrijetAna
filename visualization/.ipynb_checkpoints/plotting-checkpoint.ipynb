```python
import ROOT
from ROOT import TFile, TTree, TH1F, TCanvas, TAxis, TLegend, TTreeReader, TTreeReaderValue
ROOT.gStyle.SetOptStat(0)
```

```python
file = TFile.Open("/home/xyan13/Trijet/TrijetAna/TrijetAna/outputs/Res1ToRes2GluTo3Glu_M1-3000_R-0p5_ML_study.root")
tree = file.Get("Events")
```

```python
# variable2use = [i.GetName() for i in tree.GetListOfBranches()]
variable2use = {}
variable2use["m_jj"] = {"title": "m_{jj} [GeV]", "binning": 100, "lb": 0, "hb": 3000}
variable2use["dijet_eta"] = {"title": "#eta_{jj}", "binning": 100, "lb": 0, "hb": 3}
variable2use["dR_jj"] = {"title": "#Delta R_{jj}", "binning": 100, "lb": 0, "hb": 6}
variable2use["dEta_jj"] = {"title": "#Delta #eta_{jj}", "binning": 100, "lb": 0, "hb": 3}
variable2use["dPhi_jj"] = {"title": "#Delta #varphi_{jj}", "binning": 100, "lb": 0, "hb": 6}
variable2use["jet_eta_0"] = {"title": "#eta_{j0}", "binning": 100, "lb": 0, "hb": 3}
variable2use["jet_eta_1"] = {"title": "#eta_{j1}", "binning": 100, "lb": 0, "hb": 3}
variable2use["jet_eta_2"] = {"title": "#eta_{j2}", "binning": 100, "lb": 0, "hb": 3}
variable2use["jet_ptoverm_0"] = {"title": "p^{j0}_{T}/m_{jj}", "binning": 100, "lb": 0, "hb": 3}
variable2use["jet_ptoverm_1"] = {"title": "p^{j1}_{T}/m_{jj}", "binning": 100, "lb": 0, "hb": 3}
variable2use["dR_jj_j"] = {"title": "#Delta R_{jj-j}", "binning": 100, "lb": 0, "hb": 6}
variable2use["dEta_jj_j"] = {"title": "#Delta #eta_{jj-j}", "binning": 100, "lb": 0, "hb": 3}
variable2use["dPhi_jj_j"] = {"title": "#Delta #varphi_{jj-j}", "binning": 100, "lb": 0, "hb": 6}
variable2use["jet_ptoverM_0"] = {"title": "p^{j0}_{T}/M_{jjj}", "binning": 100, "lb": 0, "hb": 3}
variable2use["jet_ptoverM_0"] = {"title": "p^{j1}_{T}/M_{jjj}", "binning": 100, "lb": 0, "hb": 3}
variable2use["jet_ptoverM_0"] = {"title": "p^{j2}_{T}/M_{jjj}", "binning": 100, "lb": 0, "hb": 3}
variable2use["dijet_ptoverM"] = {"title": "p^{jj}_{T}/M_{jjj}", "binning": 100, "lb": 0, "hb": 3}

histo_dic = {}
for variable, param_dict in variable2use.items():
    histo_dic[variable] = {}
    for selection in ["all","match","unmatch"]:
        histo_dic[variable][selection] = TH1F(f"{variable}_{selection}",f"{variable}_{selection}",param_dict["binning"], param_dict["lb"], param_dict["hb"])
        histo_dic[variable][selection].SetCanExtend(ROOT.TH1.kYaxis)
display(histo_dic)
```

```python
for event in tree:
    for variable, hist_sel_dic in histo_dic.items():
        hist_sel_dic["all"].Fill(getattr(event, variable))
        if(event.gen_dijet_matched == 1):
            hist_sel_dic["match"].Fill(getattr(event, variable))
        else:
            hist_sel_dic["unmatch"].Fill(getattr(event, variable))
```

```python
def setRangeY(h_plot):
    global y_upper
    entries = [h_plot.GetBinContent(i) for i in range(1,h_plot.GetNbinsX())]
    max_entry = max(entries)
    print(max_entry, y_upper)
    if max_entry > y_upper:
        y_upper = max_entry * 1.1
def plot(h_plot, canvas, leg, lcolor, lstyle, xtitle, dopt, hlabel):
    global y_upper
    h_plot.Scale(1/h_plot.GetEntries())
    h_plot.SetLineWidth(2)
    h_plot.SetLineColor(lcolor)
    h_plot.SetLineStyle(lstyle)
    h_plot.Draw(dopt)
    h_plot.GetXaxis().SetTitle(xtitle)
    h_plot.GetXaxis().SetTitleOffset(1.2)
    h_plot.GetYaxis().SetTitle("Events")
    setRangeY(h_plot)
    h_plot.GetYaxis().SetRangeUser(0,y_upper)
    leg.AddEntry(h_plot,hlabel)
```

```python
import math
ROOT.disableJSVis()
nvar = len(histo_dic.keys())
nrow = math.ceil(nvar/3)
plot_list = []
for variable, hist_sel_dic in histo_dic.items():
    y_upper = 0
    c = TCanvas(variable,variable,800,600)
    leg = TLegend(0.6,0.7,0.85,0.85)
    title = variable2use[variable]["title"]
    plot(h_plot=hist_sel_dic["all"], canvas=c, leg=leg, lcolor=4, lstyle=1, xtitle=title, dopt="HIST", hlabel="all pairs")
    plot(h_plot=hist_sel_dic["match"], canvas=c, leg=leg, lcolor=2, lstyle=1, xtitle=title, dopt="HIST SAME", hlabel="matched pairs")
    plot(h_plot=hist_sel_dic["unmatch"], canvas=c, leg=leg, lcolor=ROOT.kGreen-2, lstyle=1, xtitle=title, dopt="HIST SAME", hlabel="unmatched pairs")
    c.
    plot_list.append(c)
display(plot_list)
for plot in plot_list:
    leg.Draw()
    plot.Draw()
```
