```python
sample = "M1-3000_R-0p5"
tempFile = TFile(f"{sample}_training.root","RECREATE")
outTree_sig = TTree()
outTree_bkg = TTree()
inFile = TFile(f"/home/xyan13/Trijet/TrijetAna/TrijetAna/outputs/Res1ToRes2GluTo3Glu_{sample}_ML_study.root")
inTree = inFile.Get("Events")
dijet_eta = np.empty((1), dtype="float32")
dijet_phi = np.empty((1), dtype="float32")
dR_jj = np.empty((1), dtype="float32")
dEta_jj = np.empty((1), dtype="float32")
dPhi_jj = np.empty((1), dtype="float32")
jet_eta_0 = np.empty((1), dtype="float32")
jet_phi_0 = np.empty((1), dtype="float32")
jet_ptoverm_0 = np.empty((1), dtype="float32")
jet_eta_1 = np.empty((1), dtype="float32")
jet_phi_1 = np.empty((1), dtype="float32")
jet_ptoverm_1 = np.empty((1), dtype="float32")
jet_eta_2 = np.empty((1), dtype="float32")
jet_phi_2 = np.empty((1), dtype="float32")
jet_ptoverm_2 = np.empty((1), dtype="float32")
dR_jj_j = np.empty((1), dtype="float32")
dEta_jj_j = np.empty((1), dtype="float32")
dPhi_jj_j = np.empty((1), dtype="float32")
jet_ptoverM_0 = np.empty((1), dtype="float32")
jet_ptoverM_1 = np.empty((1), dtype="float32")
jet_ptoverM_2 = np.empty((1), dtype="float32")
dijet_ptoverM = np.empty((1), dtype="float32")
gen_dijet_matched = np.empty((1), dtype="int32")

outTree_sig.Branch("dijet_eta", dijet_eta, "dijet_eta/F")
outTree_sig.Branch("dijet_phi", dijet_phi, "dijet_phi/F")
outTree_sig.Branch("dR_jj", dR_jj, "dR_jj/F")
outTree_sig.Branch("dEta_jj", dEta_jj, "dEta_jj/F")
outTree_sig.Branch("dPhi_jj", dPhi_jj, "dPhi_jj/F")
outTree_sig.Branch("jet_eta_0", jet_eta_0, "jet_eta_0/F")
outTree_sig.Branch("jet_phi_0", jet_phi_0, "jet_phi_0/F")
outTree_sig.Branch("jet_ptoverm_0", jet_ptoverm_0, "jet_ptoverm_0/F")
outTree_sig.Branch("jet_eta_1", jet_eta_1, "jet_eta_1/F")
outTree_sig.Branch("jet_phi_1", jet_phi_1, "jet_phi_1/F")
outTree_sig.Branch("jet_ptoverm_1", jet_ptoverm_1, "jet_ptoverm_1/F")
outTree_sig.Branch("jet_eta_2", jet_eta_2, "jet_eta_2/F")
outTree_sig.Branch("jet_phi_2", jet_phi_2, "jet_phi_2/F")
outTree_sig.Branch("jet_ptoverm_2", jet_ptoverm_2, "jet_ptoverm_2/F")
outTree_sig.Branch("dR_jj_j", dR_jj_j, "dR_jj_j/F")
outTree_sig.Branch("dEta_jj_j", dEta_jj_j, "dEta_jj_j/F")
outTree_sig.Branch("dPhi_jj_j", dPhi_jj_j, "dPhi_jj_j/F")
outTree_sig.Branch("jet_ptoverM_0", jet_ptoverM_0, "jet_ptoverM_0/F")
outTree_sig.Branch("jet_ptoverM_1", jet_ptoverM_1, "jet_ptoverM_1/F")
outTree_sig.Branch("jet_ptoverM_2", jet_ptoverM_2, "jet_ptoverM_2/F")
outTree_sig.Branch("dijet_ptoverM", dijet_ptoverM, "dijet_ptoverM/F")
outTree_sig.Branch("gen_dijet_matched", gen_dijet_matched, "gen_dijet_matched/I")

outTree_bkg.Branch("dijet_eta", dijet_eta, "dijet_eta/F")
outTree_bkg.Branch("dijet_phi", dijet_phi, "dijet_phi/F")
outTree_bkg.Branch("dR_jj", dR_jj, "dR_jj/F")
outTree_bkg.Branch("dEta_jj", dEta_jj, "dEta_jj/F")
outTree_bkg.Branch("dPhi_jj", dPhi_jj, "dPhi_jj/F")
outTree_bkg.Branch("jet_eta_0", jet_eta_0, "jet_eta_0/F")
outTree_bkg.Branch("jet_phi_0", jet_phi_0, "jet_phi_0/F")
outTree_bkg.Branch("jet_ptoverm_0", jet_ptoverm_0, "jet_ptoverm_0/F")
outTree_bkg.Branch("jet_eta_1", jet_eta_1, "jet_eta_1/F")
outTree_bkg.Branch("jet_phi_1", jet_phi_1, "jet_phi_1/F")
outTree_bkg.Branch("jet_ptoverm_1", jet_ptoverm_1, "jet_ptoverm_1/F")
outTree_bkg.Branch("jet_eta_2", jet_eta_2, "jet_eta_2/F")
outTree_bkg.Branch("jet_phi_2", jet_phi_2, "jet_phi_2/F")
outTree_bkg.Branch("jet_ptoverm_2", jet_ptoverm_2, "jet_ptoverm_2/F")
outTree_bkg.Branch("dR_jj_j", dR_jj_j, "dR_jj_j/F")
outTree_bkg.Branch("dEta_jj_j", dEta_jj_j, "dEta_jj_j/F")
outTree_bkg.Branch("dPhi_jj_j", dPhi_jj_j, "dPhi_jj_j/F")
outTree_bkg.Branch("jet_ptoverM_0", jet_ptoverM_0, "jet_ptoverM_0/F")
outTree_bkg.Branch("jet_ptoverM_1", jet_ptoverM_1, "jet_ptoverM_1/F")
outTree_bkg.Branch("jet_ptoverM_2", jet_ptoverM_2, "jet_ptoverM_2/F")
outTree_bkg.Branch("dijet_ptoverM", dijet_ptoverM, "dijet_ptoverM/F")
outTree_bkg.Branch("gen_dijet_matched", gen_dijet_matched, "gen_dijet_matched/I")

n_all = inTree.GetEntries()
array_ismatch = inTree.AsMatrix(columns=["gen_dijet_matched"])
np.squeeze(array_ismatch)
n_matched = sum(array_ismatch)[0]
n_unmatched = n_all - n_matched
print(n_all)
ratio = n_matched / n_unmatched
from ROOT import TRandom3
rand_gen = TRandom3()
count_unmatch = 0
for i, event in enumerate(inTree):
    if i%10000 == 0:
        print("Processing: ",i)
    if(event.gen_dijet_matched == 0):
        random = rand_gen.Uniform(0,1)
        if random > ratio: continue
    dijet_eta[0] = event.dijet_eta
    dijet_phi[0] = event.dijet_phi
    dR_jj[0] = event.dR_jj
    dEta_jj[0] = event.dEta_jj
    dPhi_jj[0] = event.dPhi_jj
    jet_eta_0[0] = event.jet_eta_0
    jet_phi_0[0] = event.jet_phi_0
    jet_ptoverm_0[0] = event.jet_ptoverm_0
    jet_eta_1[0] = event.jet_eta_1
    jet_phi_1[0] = event.jet_phi_1
    jet_ptoverm_1[0] = event.jet_ptoverm_1
    jet_eta_2[0] = event.jet_eta_2
    jet_phi_2[0] = event.jet_phi_2
    jet_ptoverm_2[0] = event.jet_ptoverm_2
    dR_jj_j[0] = event.dR_jj_j
    dEta_jj_j[0] = event.dEta_jj_j
    dPhi_jj_j[0] = event.dPhi_jj_j
    jet_ptoverM_0[0] = event.jet_ptoverM_0
    jet_ptoverM_1[0] = event.jet_ptoverM_1
    jet_ptoverM_2[0] = event.jet_ptoverM_2
    dijet_ptoverM[0] = event.dijet_ptoverM
    gen_dijet_matched[0] = event.gen_dijet_matched
    if(event.gen_dijet_matched == 0):
        outTree_bkg.Fill()
    else:
        outTree_sig.Fill()
tempFile.Write()
tempFile.Close()
```

```python
from ROOT import TMVA
TMVA.Tools.Instance()
outFile = TFile("temp.root","RECREATE")
factory = TMVA.Factory("MVAnalysis", outFile,"!V:!Silent:Color:DrawProgressBar:Transformations=I;D;P;G,D:AnalysisType=Classification")
dataloader = TMVA.DataLoader(sample)
dataloader.AddSignalTree(outTree_sig)
dataloader.AddBackgroundTree(outTree_bkg)
dataloader.AddVariable("dijet_eta", 'F')
dataloader.AddVariable("dijet_phi", 'F')
dataloader.AddVariable("dR_jj", 'F')
dataloader.AddVariable("dEta_jj", 'F')
dataloader.AddVariable("dPhi_jj", 'F')
dataloader.AddVariable("jet_eta_0", 'F')
dataloader.AddVariable("jet_phi_0", 'F')
dataloader.AddVariable("jet_ptoverm_0", 'F')
dataloader.AddVariable("jet_eta_1", 'F')
dataloader.AddVariable("jet_phi_1", 'F')
dataloader.AddVariable("jet_ptoverm_1", 'F')
dataloader.AddVariable("jet_eta_2", 'F')
dataloader.AddVariable("jet_phi_2", 'F')
dataloader.AddVariable("jet_ptoverm_2", 'F')
dataloader.AddVariable("dR_jj_j", 'F')
dataloader.AddVariable("dEta_jj_j", 'F')
dataloader.AddVariable("dPhi_jj_j", 'F')
dataloader.AddVariable("jet_ptoverM_0", 'F')
dataloader.AddVariable("jet_ptoverM_1", 'F')
dataloader.AddVariable("jet_ptoverM_2", 'F')
dataloader.AddVariable("dijet_ptoverM", 'F')
cut_s = ROOT.TCut()
cut_b = ROOT.TCut()
dataloader.PrepareTrainingAndTestTree(cut_s,cut_b,"nTrain_Signal=0:nTrain_Background=0:SplitMode=Random:NormMode=NumEvents:!V" )
factory.BookMethod(dataloader, TMVA.Types.kBDT, "BDT", 
                   "!H:!V:NTrees=500:MinNodeSize=2.5%:MaxDepth=3:BoostType=AdaBoost:AdaBoostBeta=0.5:SeparationType=GiniIndex:nCuts=20:PruneMethod=CostComplexity:PruneStrength=-1" )
factory.TrainAllMethods()
factory.TestAllMethods()
factory.EvaluateAllMethods()
outFile.Close()
```
